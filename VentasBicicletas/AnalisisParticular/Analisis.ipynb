# Análisis exploratorio de comportamiento de compra inusual.

Este script tiene como objetivo identificar patrones detrás de compras múltiples en una tienda de bicicletas, con el fin de replicarlos y potenciar las ventas. El foco está en entender por qué cuatro clientes adquirieron dos unidades en el año 2015.

## Enfoque del análisis.
Se utilizaron gráficos de caja para evaluar si variables como ``ingreso_anual`` y ``numero_hijos`` presentaban comportamientos atípicos. Además, se revisó la estadística descriptiva y se exploraron posibles factores como fechas de cumpleaños o ubicación de compra.

## Hallazgos clave.

- Los clientes que compraron dos bicicletas no presentan valores atípicos en ingreso ni número de hijos.

- La hipótesis de compra por cumpleaños fue descartada.

- Se identificó que tres de los cuatro clientes realizaron su compra en la tienda AW Breña, lo que sugiere una posible influencia externa como promociones o incentivos específicos.

# Conclusión.

Aunque los datos disponibles no permiten establecer una causa definitiva, se recomienda investigar si la tienda AW Breña ofreció promociones especiales que motivaron la compra de más de una unidad. Este tipo de análisis puede ser clave para diseñar estrategias comerciales más efectivas.

## Gráfica de Cajas

```python
def boxplot(data, target):
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=data, x='genero', y=target, palette='Set2')
    plt.title(f'Distribución de {target} por género')
    plt.xlabel('Género')
    plt.ylabel(target)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
boxplot(df, 'ingreso_anual')
boxplot(df, 'numero_hijos')
```
